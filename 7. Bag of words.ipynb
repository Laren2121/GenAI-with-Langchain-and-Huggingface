```python
import pandas as pd 
messages = pd.read_csv('SpamClassifier-master/SMSSpamCollection.txt', sep = '\t', names = ["label", "message"])
```

```python
messages
```

```python
#Data cleaning and preprocessing
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

porter_stemmer = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [porter_stemmer.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Apply bag of words
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(max_features = 100, binary = True)
```

```python
X = count_vectorizer.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
X
```

### Lemmatization Method

```python
import re
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords 
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []

for i in range(0, len(messages)):
    review =  re.sub('[^a-zA-Z]', ' ' , messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word, pos ='v') for word in review if review not in stopwords.words('english')]
    review = ' '.join(review)
    #print(review)
    corpus.append(review)
    #print(review)
```

```python
corpus
```

```python
## Create a Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(max_features = 100, binary = True)
```

```python
X = count_vectorizer.fit_transform(corpus).toarray()
```

```python
## Creating Bag OF Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 500, binary = True, ngram_range = (2, 3))
X = cv.fit_transform(corpus).toarray()
```
